```python
s = "python python is good good"
l = s.split()
r = []
print(l)

for i in l:
    if not i in r:
        r.append(i)
        
print(" ".join(r))
```

```python
def adder(k, v):
    result = {}
    
    for i in range(len(k)):
        result[k[i]] = v[i]
    return result

print(adder([" name" ," age" ," city" ], [" ali" ,25," tehran" ]))
```
