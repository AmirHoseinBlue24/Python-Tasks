```python
import re

p = "Ali@1234"

if re.findall(r"\w", p) and re.findall(r"\d", p) and re.findall(r"\D", p) and re.findall(r"\S", p) and re.findall(r"\W", p):
    print("Strong")
else:
    print("Weak")
```

```python
s = "hello"
d = {}

for i in range(len(s)):
    w = s[i]
    if w not in d:
        d[w] = []
    d[w].append(i)

print(d)
```
