```python
import re

s = "apple banana apple orange banana apple"
l = re.split('\s', s)
d = {}
print (l)

for i in l:
    d[i] = d.get(i, 0)+1

print(d)
```

```python
s =  "javascript is more powerful than java"
if s != "":
    l = s.split(' ')
    print(max(l, key=len))
print(l)
```

```python
import re

s = "h@e#l$l%o"
c = re.findall('\w', s)
t = ""

for i in c:
    t += i

print(t)
```
